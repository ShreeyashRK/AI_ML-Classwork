```python
cost_max=max(df['Cost'])
cost_min=min(df['Cost'])
area_min=min(df['Area'])
area_max=max(df['Area'])
print(cost_min,cost_max,area_min,area_max)
k= cost_max-cost_min
l=area_max-area_min
```

```python
d=[]
for i in df['Cost']:
    d.append((i-cost_min)/(k))
print(d)
df['Cost']=d
```
