```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['B', 'C', 'D', 'E']
}
```

```python
# DFS
def dfs(st, goal, graph, vis=None):

    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False


dfs('A', 'F', graph)
```

```python
# BFS
def bfs(graph, st, goal):

    vis = set()
    queue = [st]
    path = []

    vis.add(st)

    while queue:

        node = queue.pop(0)
        path.append(node)

        if node == goal:
            return path

        for neighbour in graph[node]:
            if neighbour not in vis:
                vis.add(neighbour)
                queue.append(neighbour)

    return path


print(bfs(graph, 'A', 'F'))
```
