```python
import networkx as nwx
graph=nwx.Graph()
while True:
    ch=int(input("Want to insert nodes and enter edge press? 1 for yes, 2 for no"))
    if ch==1:
        n1=input().strip().upper()
        n2=input().strip().upper()
        wt=int(input("Enter weight"))
        graph.add_edge(n1,n2,weight=wt)
    else:
        break
    for node,neighbours in graph.adj.items():
        print(node,neighbours)
```

```python

st=input().strip().upper()
goal=input().strip().upper()
heuristic_values={}
heuristic_values[goal]=0
for node in graph.nodes():
    if node!=goal:
        heuristic_val=int(input(f"Enter the heuristic value from node {node} to goal node {goal} "))
        heuristic_values[node]=heuristic_val
print(heuristic_values)
```

```python
heuristic_values
```

```python
def get_heuristic(u,v):
    return heuristic_values[u]
```

```python
path=nwx.astar_path(graph,st,goal,heuristic=get_heuristic,weight='weight')
"->".join(path)
```

```python
total_cost=nwx.astar_path_length(graph,st,goal,heuristic=get_heuristic,weight='weight')
print(total_cost)
```
