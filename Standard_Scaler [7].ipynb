```python
import pandas as pd
df=pd.read_csv("healthcare.csv")
df
```

```python
df['Age'].isnull().sum()
```

```python
df['BloodPressure'].isnull().sum()
```

```python
age_mean=df['Age'].mean()
bp_mean=df['BloodPressure'].mean()
age_std=df['Age'].std()
bp_std=df['BloodPressure'].std()


a=[]
bp=[]
for i in df['Age']:
    a.append((i-age_mean)/age_std)
for j in df['BloodPressure']:
    bp.append((j-bp_mean)/bp_std)

print("Age column is:")
for i in a:
    print(i)
print("\n\n")
print("Blood Pressure column is: ")
for j in bp:
    print(j)
```

```python
import numpy as np
a=np.array(a)
bp=np.array(bp)
```

```python
print("New age mean is: ",np.mean(a))
print("New bp mean is: ",np.mean(bp))
print("New age std is: ",np.std(a))
print("New bp std is: ",np.std(bp))
```
