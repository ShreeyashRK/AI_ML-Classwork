```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(np.concatenate([arr1, arr2]))
```

```python
a = np.array([[1,2,3,4], [1,2,3,4]])
b = np.array([[5,6,7,8], [5,6,7,8]])

print(np.concatenate((a, b), axis=0))
print(np.concatenate((a, b), axis=1))
```
